```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_input = '../data/'
path_output = '../figures/'
```

```python
ds_gs = xr.open_dataset(path_input + 'GS_along_track_spec.nc')
ds_agulhas = xr.open_dataset(path_input + 'Agulhas_along_track_spec.nc')
ds_ccs = xr.open_dataset(path_input + 'CCS_along_track_spec.nc')
ds_acc1 = xr.open_dataset(path_input + 'ACC1_along_track_spec.nc')
ds_acc2 = xr.open_dataset(path_input + 'ACC2_along_track_spec.nc')
ds_med = xr.open_dataset(path_input + 'MED_along_track_spec.nc')
```

```python
fac = (2*np.pi)/1000
scale_oi = 20 # in km
fig, axes = plt.subplots(ncols = 6, figsize = (9, 3), sharey = True)
ax = axes[0]
ax.loglog(ds_gs.wavenumber/fac, ds_gs.spec_raw * fac, label = 'L3 product')
ax.loglog(ds_gs.wavenumber/fac, ds_gs.spec_oi * fac, label = 'Interpolated SWOT')
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_gs.spec_oi * fac), np.amax(ds_gs.spec_oi * fac)])
ax.set_xlim([1/300, 1/3])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('Gulf Stream')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)
fig.legend(bbox_to_anchor=[.3, 1.2])

ax = axes[1]
ax.loglog(ds_agulhas.wavenumber/fac, ds_agulhas.spec_raw * fac)
ax.loglog(ds_agulhas.wavenumber/fac, ds_agulhas.spec_oi * fac)
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_agulhas.spec_oi * fac), np.amax(ds_agulhas.spec_oi * fac)])
ax.set_xlim([1/300, 1/3])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('Agulhas')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)
ax = axes[2]
ax.loglog(ds_ccs.wavenumber/fac, ds_ccs.spec_raw * fac)
ax.loglog(ds_ccs.wavenumber/fac, ds_ccs.spec_oi * fac)
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_ccs.spec_oi * fac), np.amax(ds_ccs.spec_oi * fac)])
ax.set_xlim([1/300, 1/2])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('CCS')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)

ax = axes[3]
ax.loglog(ds_acc1.wavenumber/fac, ds_acc1.spec_raw * fac)
ax.loglog(ds_acc1.wavenumber/fac, ds_acc1.spec_oi * fac)
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_acc1.spec_oi * fac), np.amax(ds_acc1.spec_oi * fac)])
ax.set_xlim([1/300, 1/3])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('ACC1')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)

ax = axes[4]
ax.loglog(ds_acc2.wavenumber/fac, ds_acc2.spec_raw * fac)
ax.loglog(ds_acc2.wavenumber/fac, ds_acc2.spec_oi * fac)
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_acc2.spec_oi * fac), np.amax(ds_acc2.spec_oi * fac)])
ax.set_xlim([1/300, 1/3])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('ACC2')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)


ax = axes[5]
ax.loglog(ds_med.wavenumber/fac, ds_med.spec_raw * fac)
ax.loglog(ds_med.wavenumber/fac, ds_med.spec_oi * fac)
ax.plot([1/scale_oi, 1/scale_oi], [10**(-10), 30**1], 'k')
ax.set_ylim([np.amin(ds_med.spec_oi * fac), np.amax(ds_med.spec_oi * fac)])
ax.set_xlim([1/300, 1/3])
ax.set_ylim([10**(-10), 30**1])
ax.set_title('MED')
ax.grid(visible=True, which='major', color='k', linestyle='-', alpha = .4)
ax.grid(visible=True, which='minor', color='k', linestyle='--', alpha = .2)

fig.supxlabel('Wavenumber [cpkm]')
fig.supylabel('PSD [$m^{2}/cpkm]$')

plt.tight_layout()
ax_inset1 = fig.add_axes([.68, 1.2, .12, .5])
ax_inset1.pcolormesh(ds_gs.longitude, ds_gs.latitude, ds_gs.ssha_raw, cmap = 'RdBu_r')
ax_inset1.set_title('SWOT L3')
ax_inset1.set_xlabel('Longitude [$^{\circ}$E]')
ax_inset1.set_ylabel('Latitude [$^{\circ}$N]')

ax_inset2 = fig.add_axes([.85, 1.2, .13, .5])
p1 = ax_inset2.pcolormesh(ds_gs.longitude, ds_gs.latitude, ds_gs.ssha_oi, cmap = 'RdBu_r')
ax_inset2.set_title('Interpolated')
ax_inset2.set_xlabel('Longitude [$^{\circ}$E]')
plt.colorbar(p1, ax = ax_inset2, label = 'SSHA [m]')
plt.savefig(path_output + 'Fig_S1.png', dpi = 300, bbox_inches = 'tight')
```
