```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_input = '/Users/gwen29360/Desktop/POSTDOC/LEGOS/SWOT/data_composite_paper/'
path_output = '/Users/gwen29360/Desktop/POSTDOC/LEGOS/SWOT/figure_composite_paper/'
```

```python


def binned_scatter(x, y, nbins=20):
    x = np.asarray(x)
    y = np.asarray(y)

    bins = np.linspace(np.nanpercentile(x, 1), np.nanpercentile(x, 99), nbins + 1)
    bin_centers = 0.5 * (bins[:-1] + bins[1:])

    digitized = np.digitize(x, bins) - 1

    y_mean = np.full(nbins, np.nan)
    y_std  = np.full(nbins, np.nan)

    for i in range(nbins):
        mask = digitized == i
        if np.any(mask):
            y_mean[i] = np.nanmean(y[mask])
            y_std[i]  = np.nanstd(y[mask])

    return bin_centers, y_mean, y_std

cases = ['GS2', 'AGULHAS', 'CCS', 'ACC1_bis', 'ACC2', 'MED']
r_max_vec_swot = []
v_max_vec_swot = []

for case in cases:

    ds_swot = xr.open_dataset(path_input + f'DYNED_{case}_SWOT_L3.nc')
    ds_apparaillage = xr.open_dataset(path_input + f'appareillage_{case}_atlas.nc')

    indices_swot_appareille = ds_apparaillage.matched_pairs.sel(platforms = 1).values

    ds_AMEDA_SWOT_appareiller = ds_swot.isel(obs = indices_swot_appareille)

    r_max_vec_swot.extend(ds_AMEDA_SWOT_appareiller.r_max.values)
    v_max_vec_swot.extend(ds_AMEDA_SWOT_appareiller.v_max.values)


r_max_vec_swot = np.array(r_max_vec_swot)
v_max_vec_swot = np.array(v_max_vec_swot)



fig, ax = plt.subplots(figsize = (3, 4))
# ax.scatter(r_max_vec_swot, v_max_vec_swot)
bin_centers, y_mean, y_std = binned_scatter(r_max_vec_swot, v_max_vec_swot, nbins=10)
ax.errorbar(bin_centers, y_mean, yerr = y_std, capsize = 3, color = 'r', label = 'bin averaged')
ax.scatter(r_max_vec_swot, v_max_vec_swot, color = 'k', alpha = .05)
ax.scatter(r_max_vec_swot[10], v_max_vec_swot[10], color = 'k', alpha = .5, label = 'Individual eddies')

ax.set_xlabel('R$_{max}$ [km]')
ax.set_ylabel('V$_{max}$ [m/s]')
ax.set_title('All SWOT Eddies')
plt.savefig(path_output + 'FigS4.png', dpi = 300, bbox_inches = 'tight')
```
